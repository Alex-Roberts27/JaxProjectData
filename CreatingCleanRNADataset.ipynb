```python
# import the stack!
import pandas as pd
import numpy as np
from scipy.stats import chisquare
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
# typical Data Cleaning stack
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler,OrdinalEncoder,OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
file1 = pd.read_excel('rawcountdata.xlsx')
file1.to_csv('rawcounts.csv', index=False)
print(file1.info())
```

```python
# how much missing data is there in each of these columns?
print(file1.isnull().sum())
```

```python
# drop duplicate rows
dup = file1.duplicated(keep='first')
dropped = file1.loc[dup]

if not dropped.empty:
    print(f"Found {len(dropped)} duplicate rows.")
    print(dropped['rsID'].unique()[0]) # Still useful if duplicates are found to show an example.
else:
    print("No duplicate rows found.")
```

```python
print(file1.shape)
```

```python
count_columns = ['Ctrl_1_RAW3', 'Ctrl_2_RAW3', 'Ctrl_3_RAW3', 'Inf_1_RAW3', 'Inf_2_RAW3', 'Inf_3_RAW3', 'Ctrl_1_RAW14', 'Ctrl_2_RAW14', 'Ctrl_3_RAW14', 'Inf_1_RAW14', 'Inf_2_RAW14', 'Inf_3_RAW14']
# Keep only rows where ALL samples have counts > 0
file1_filtered = file1[(file1[count_columns] > 0).all(axis=1)]
file1__filtered = file1[(file1[count_columns] > 0).any(axis=1)]
```

```python
print(file1_filtered.shape)
```

```python
file1_filtered.to_csv('filteredrawcounts.csv', index=False)
```

```python
# 1. Load file
df = pd.read_csv('filteredrawcounts.csv')

# 2. Define your threshold
threshold = 20

# 3. Filter the data
# Think about which comparison operator (>, <, ==) fits "above a certain value"
filtered_df = df[(df[count_columns] > threshold).any(axis=1)]

# 4. Save the cleaned data back to a CSV
filtered_df.to_csv('thresholdrawcounts.csv', index=False)
```
